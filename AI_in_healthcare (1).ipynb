**ENCODING** FOR BETTER ACCURACY

```python
df.Disease = [0 if value == "Healthy" else 1 for value in df.Disease]
df_tst.Disease = [0 if value == "Healthy" else 1 for value in df_tst.Disease]
```

**SPLITTING DATA INTO TEST ANS TRAIN**

```python
x_train = df.drop('Disease', axis=1)
y_train = df['Disease']
x_test = df_tst.drop('Disease', axis=1)
y_test = df_tst['Disease']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

**RANDOM FOREST CLASSIFIER**

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

**GRADIENT BOOSTING**

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier(n_estimators=100, learning_rate=0.1)
gb.fit(x_train,y_train)
```

```python
gb.score(x_test,y_test)
```

**SUPPORT VECTOR MACHINE**

```python
from sklearn.svm import SVC
svc=SVC(C= 100, gamma=0.1, kernel='rbf')
svc.fit(x_train,y_train)
```

```python
svc.score(x_test,y_test)
```

**LOGISTIIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
l=[RandomForestClassifier(), GradientBoostingClassifier(), SVC(), LogisticRegression()]
result = {}
for i in l :
  i.fit(x_train, y_train)
  result[i] = i.score(x_test, y_test)
```

```python
print(result)
```

**HEATMAPS**

```python
y_pred_rf=rf.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
sns.heatmap(confusion_matrix(y_test,y_pred_rf), annot=True, fmt='d',
 cmap='Blues')
#print(classification_report(y_test,y_pred_rf))
```

```python
y_pred_gb=gb.predict(x_test)
sns.heatmap(confusion_matrix(y_test,y_pred_gb), annot=True, fmt='d',
 cmap='Blues')
```
